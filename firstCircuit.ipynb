```python
from qiskit import *
```

```python
devre=QuantumCircuit(2,2)
```

```python
devre.h(0)
```

```python
%matplotlib inline
```

```python
devre.cx(0,1)
```

```python
devre.draw(output="mpl")
```

```python
devre.measure([0,1],[0,1])
devre.draw(output="mpl")
```

```python
simulator=Aer.get_backend("qasm_simulator")
```

```python
result=execute(devre,backend=simulator).result()
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(devre))
```
