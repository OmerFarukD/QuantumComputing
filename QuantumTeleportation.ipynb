```python
from qiskit import*
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2)
circuit.barrier()
circuit.draw(output="mpl")
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
circuit.draw(output="mpl")
```

```python
circuit.measure([0,1],[0,1])
circuit.draw(output="mpl")
```

```python
circuit.cx(1,2)
circuit.cz(0,2)
circuit.measure(2,2)
```

```python
circuit.draw(output="mpl")
```

```python
simulator=Aer.get_backend("qasm_simulator")
```

```python
job=execute(circuit,backend=simulator)
result=job.result()
count=result.get_counts()
```

```python
plot_histogram(count)
```
