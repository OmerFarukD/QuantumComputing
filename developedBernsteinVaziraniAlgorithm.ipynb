```python
from qiskit import *
%matplotlib inline
```

```python
secretNumber="1000101001010111010010010"
devre=QuantumCircuit(len(secretNumber)+1,len(secretNumber))
```

```python
devre.h(range(len(secretNumber)))
devre.x(len(secretNumber))
devre.h(len(secretNumber))
devre.barrier()
```

```python
for index, deger in enumerate(secretNumber[::-1]):  #[::-1] stringi tersten yazar başka bir kullanımı ise reversed(secretNumber)
    if deger=="1":
        devre.cx(index,len(secretNumber))
devre.barrier()
        
```

```python
devre.h(range(len(secretNumber)))
```

```python
devre.measure(range(len(secretNumber)),range(len(secretNumber)))
```

```python
devre.draw(output="mpl")
```

```python
qasm_sim=Aer.get_backend("qasm_simulator")
```

```python
qasm_job=execute(devre,backend=qasm_sim,shots=1).result()
```

```python
counts=qasm_job.get_counts()
```

```python
print(counts)
```
